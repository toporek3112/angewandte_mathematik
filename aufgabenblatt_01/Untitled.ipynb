a) Vector (-3, 1)

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure and an axis
fig, ax = plt.subplots()

# A sample vector (x_position, y_position, x_direction, y_direction)
v_x = np.array([-3, 1])
v_y = np.array([-4, 2])

# Using quiver to plot the vector
ax.quiver(*v_x, scale=1, scale_units='xy', angles='xy')
ax.quiver(*v_y, scale=1, scale_units='xy', angles='xy')

# Setting the limits and labels
ax.set_xlim(-4, 1)
ax.set_ylim(0, 2)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# Showing the plot
plt.grid()
plt.show()

sp = 0
for i in range(len(v_x)):
    sp += v_x[i] * v_y[i]

print(f'Standartskalarproduct: {sp}')
```

BSP 9.3

```python
import matplotlib.pyplot as plt
import numpy as np

def get_point(ab, a, b):
    #FS-> = S - F
    if (len(a) == 0):
       a = b - ab
       return a 
    elif (len(b) == 0):
        b = ab + a
        return b
    
# Definitions
point_origin = np.array([0, 0])
vector = np.array([-3, 1])

# Vector PQ->
point_p = np.array([2, 1])
point_q = get_point(vector, point_p, [])

# Vector RS->
point_s = np.array([0, 1])
point_r = get_point(vector, [], point_s)

# Creating the plot
plt.figure(figsize=(8, 6))
plt.quiver(*point_origin, *vector, color='r', scale=1, scale_units='xy', angles='xy', label=f'Ortsvektor {vector}')
plt.quiver(*point_p, *vector, color='b', scale=1, scale_units='xy', angles='xy', label=f'Vektor PQ {vector} von P {point_p}')
plt.quiver(*point_r, *vector, color='g', scale=1, scale_units='xy', angles='xy', label=f'Vektor RS {vector} von R {point_r}')

# # Marking points
plt.scatter(*point_origin, color='black', label=f'Origin {point_origin}')
plt.scatter(*point_p, color='green', label=f'Point {point_p}')
plt.scatter(*point_q, color='purple', label=f'Point Q {point_q}')
plt.scatter(*point_r, color='red', label=f'Point R {point_r}')
plt.scatter(*point_s, color='yellow', label=f'Point S {point_s}')

# # Adding labels, grid, legend, and setting limits
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.title('Veranschaulichung des Vektors (-3, 1) im R2')

# Displaying the plot
plt.show()
```

L√§nge eines Vectors

```python
def get_vector_lenght(vector):
  res = np.square(vector).sum()
  return np.sqrt(res)

print(get_vector_lenght(vector))
print(np.linalg.norm(vector))

def get_point_distance(a, b):
  res = np.square(a - b).sum()
  return np.sqrt(res)

print(get_point_distance(point_p, point_q))
```

```python
# Defining the vectors
a1 = (2, 0, 1)
a2 = (1, 2, 0)
a3 = (0, 0, 2)

# Creating a 3D plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Adding the vectors to the plot
ax.quiver(0, 0, 0, a1[0], a1[1], a1[2], color='r', label='a1')
ax.quiver(0, 0, 0, a2[0], a2[1], a2[2], color='g', label='a2')
ax.quiver(0, 0, 0, a3[0], a3[1], a3[2], color='b', label='a3')

# Setting the plot limits
ax.set_xlim([0, 3])
ax.set_ylim([0, 3])
ax.set_zlim([0, 3])

# Adding labels and title
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D vector plot')

# Adding a legend
ax.legend()

# Showing the plot
plt.show()
```
